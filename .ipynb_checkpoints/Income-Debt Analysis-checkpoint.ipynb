```python
#importing useful libraries
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/2021_data_truncated.csv", low_memory=False)
df = df.loc[:, ['debt_to_income_ratio','action_taken']]
df.head(25)
```

```python
import numbers
df = df.dropna()
df1 = df[df['debt_to_income_ratio'].str.isnumeric()]
df2 = df[df['debt_to_income_ratio'].str.isnumeric() == False]
df1[['debt_to_income_ratio']] = df1[['debt_to_income_ratio']].apply(pd.to_numeric) 
df1.head(25)
```

```python
df2.head(25)
```

```python
# Place for testing out values
# print(len(df1)," ",len(df2))
# df1.describe()
df2['debt_to_income_ratio'].unique()
```

```python
# Numbers will represent how many people in each debt_to_income_ratio category got accepted and how many got declined (df1)

df3 = df1[(df1["debt_to_income_ratio"] >= 35) & (df1["debt_to_income_ratio"] < 40)]
num1acc = len(df3[df3["action_taken"] == 1])
num1rej = len(df3[df3["action_taken"] == 3])
df3 = df1[(df1["debt_to_income_ratio"] >= 40) & (df1["debt_to_income_ratio"] < 45)]
num2acc = len(df3[df3["action_taken"] == 1])
num2rej = len(df3[df3["action_taken"] == 3])
df3 = df1[(df1["debt_to_income_ratio"] >= 45) & (df1["debt_to_income_ratio"] < 50)]
num3acc = len(df3[df3["action_taken"] == 1])
num3rej = len(df3[df3["action_taken"] == 3])
print(num1acc," ",num1rej," ",num2acc," ",num2rej," ",num3acc," ",num3rej," ")
# df1.dtypes
```

```python
#df1 graph
labels = ['[35,40)', '[40,45)', '[45,50)']
accepted = [num1acc,num2acc,num3acc]
rejected = [num1rej,num2rej,num3rej]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, accepted, width, label='Accepted') #The x axis starts at -1
rects2 = ax.bar(x + width/2, rejected, width, label='Rejected')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Applicants')
ax.set_title('Mortages accepted and declined based of income-to-debt-ratio group')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
# Numbers will represent how many people in each debt_to_income_ratio category got accepted and how many got declined (df2)

df3 = df2[(df2["debt_to_income_ratio"] == '<20%')]
num1acc = len(df3[df3["action_taken"] == 1])
num1rej = len(df3[df3["action_taken"] == 3])
df3 = df2[(df2["debt_to_income_ratio"] == '20%-<30%')]
num2acc = len(df3[df3["action_taken"] == 1])
num2rej = len(df3[df3["action_taken"] == 3])
df3 = df2[(df2["debt_to_income_ratio"] == '30%-<36%')]
num3acc = len(df3[df3["action_taken"] == 1])
num3rej = len(df3[df3["action_taken"] == 3])
df3 = df2[(df2["debt_to_income_ratio"] == '50%-60%')]
num4acc = len(df3[df3["action_taken"] == 1])
num4rej = len(df3[df3["action_taken"] == 3])
df3 = df2[(df2["debt_to_income_ratio"] == '>60%')]
num5acc = len(df3[df3["action_taken"] == 1])
num5rej = len(df3[df3["action_taken"] == 3])
df3 = df2[(df2["debt_to_income_ratio"] == 'Exempt')]
num6acc = len(df3[df3["action_taken"] == 1])
num6rej = len(df3[df3["action_taken"] == 3])
#['<20%', '20%-<30%', '>60%', '50%-60%', '30%-<36%', 'Exempt']
# print(num1acc," ",num1rej," ",num2acc," ",num2rej," ",num3acc," ",num3rej," ")
# # df1.dtypes
```

```python
#df1 graph
labels = ['[0,20)', '[20,30)', '[30,36)','[50,60)', 'Above 60','Excempt']
accepted = [num1acc,num2acc,num3acc,num4acc,num5acc,num6acc]
rejected = [num1rej,num2rej,num3rej,num4rej,num5rej,num6rej]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, accepted, width, label='Accepted') #The x axis starts at -1
rects2 = ax.bar(x + width/2, rejected, width, label='Rejected')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Number of Applicants')
ax.set_title('Mortages accepted and declined based of income-to-debt-ratio group')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
